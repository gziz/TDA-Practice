## Intro

```python
def get_preimg(f, lim_inf, lim_sup, domain):
    preimage = []
    for x in domain:
        if f(x) >= lim_inf and f(x) <= lim_sup:
            preimage.append(x)
            
    return preimage
```

```python
# Definir el dominio (valores en x)
X = np.linspace(0, 2*np.pi, 10000)

# Definir los limites de la imagen (valores en y)
y_lim_inf, y_lim_sup = [-0.5, 0.5]
```

```python
look_up = {x : np.sin(x) for x in X}
f = lambda x: look_up[x]
```

```python
# Dado los limites de la imagen (valores en y) y el dominio.
#Â Encontrar la pre imagen (valores en x)
pre_img = get_preimg(f, y_lim_inf, y_lim_sup, X)
```

```python
y = [f(x) for x in pre_img]
plt.scatter(x=pre_img, y=y);
```

## Ejercicio 1

```python
def greater_or_equal(a, b):
    return a >= b
def greater(a, b):
    return a > b
def smaller(a, b):
    return a < b
def smaller_or_equal(a, b):
    return a <= b

interval_type_map = {
    "[]": [greater_or_equal, smaller_or_equal],
    "(]": [greater, smaller_or_equal],
    "()": [greater, smaller],
    "[)": [greater_or_equal, smaller]
}
```

```python
def get_preimg_v2(f, lim_inf: float, lim_sup: float, domain: float, interval_type: str):

    if interval_type not in interval_type_map:
        raise ValueError("Invalid interval type")
    
    func_lim_inf, func_lim_sup = interval_type_map[interval_type]

    preimage = []
    for x in domain:
        if func_lim_inf(f(x), lim_inf) and func_lim_sup(f(x), lim_sup):
            preimage.append(x)
            
    return preimage
```

```python
y_lim_inf, y_lim_sup = [-0.5, 0.5]
pre_img = get_preimg_v2(f, y_lim_inf, y_lim_sup, X, "(]")
y = [f(x) for x in pre_img]
plt.scatter(x=pre_img, y=y);
```

```python
y_lim_inf, y_lim_sup = [-1, np.sin(np.pi/3)]
pre_img = get_preimg_v2(f, y_lim_inf, y_lim_sup, X, "[)")
y = [f(x) for x in pre_img]
plt.scatter(x=pre_img, y=y);
```

```python
y_lim_inf, y_lim_sup = [np.sin(np.pi/4), 1]
pre_img = get_preimg_v2(f, y_lim_inf, y_lim_sup, X, "(]")

y = [f(x) for x in pre_img]
plt.scatter(x=pre_img, y=y);
```
